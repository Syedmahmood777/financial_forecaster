```python
#Importign all the uselful libraries
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Importing Excel files
mon_pd=pd.read_csv(r'C:\Users\WELCOME\Desktop\financial_forecaster\monthly_transactions.csv',index_col="Date",parse_dates=True)
mon_depo=pd.read_csv(r'C:\Users\WELCOME\Desktop\financial_forecaster\monthly_deposits.csv',index_col="Date",parse_dates=True)
mon_depo.head()
```

```python
#Kernel Density Esstimate plot is done for Monthly Transactions using Seaborn 
a4_dims = (9, 6)
fig, ax = plt.subplots(figsize=a4_dims)
sns.color_palette("crest", as_cmap=True)
sns.kdeplot(mon_pd['Withdrawls'],fill=True)
plt.show()
```

```python
#Kernel Density Esstimate plot is done for Monthly Deposits using Seaborn 
a4_dims = (9, 6)
fig, ax = plt.subplots(figsize=a4_dims)
sns.color_palette("crest", as_cmap=True)
sns.kdeplot(mon_depo['Deposits'],fill=True)
plt.show()
```

```python
#Histogram Plot for Monthly Withdrawls with/wihtout UPI
fig, axs = plt.subplots(1, 2, figsize=(11, 6))

sns.set(style="darkgrid")
sns.histplot(data=mon_pd, x="Yes", color="skyblue", label="Payment With UPI", kde=True,ax=axs[0]).set(title="Monthly Withdrawls using UPI")
sns.histplot(data=mon_pd, x="No", color="red", label="Payment Without UPI", kde=True,ax=axs[1]).set(title="Monthly Withdrawls without using UPI")

plt.show()
```

```python
#Histogram for Monthly Deposist and Withdrawls

sns.set(style="darkgrid")

fig, axs = plt.subplots(1, 2, figsize=(11, 6))
sns.histplot(data=mon_pd, x="Total", color="navy",kde=True,bins=5,ax=axs[0]).set(title="Total Withdrawls in a month")

sns.histplot(data=mon_depo, x="Total", color="skyblue", kde=True,bins=5,ax=axs[1]).set(title="Total Deposits in a month")
plt.show()
```

```python
# Gaussian KDE is plotted for monthly withdrawls
from scipy.stats import gaussian_kde
from numpy import linspace

plt.rcParams["figure.figsize"]=12,8

sns.kdeplot(data=mon_pd, x="Withdrawls",  fill=True, alpha=1)


kde = gaussian_kde(mon_pd.Withdrawls)
x_range = linspace(min(mon_pd.Withdrawls), max(mon_pd.Withdrawls), len(mon_pd.Withdrawls))


sns.lineplot(x=x_range*-1, y=kde(x_range) * -1, color='orange') 
plt.fill_between(x_range*-1, kde(x_range) * -1, color='orange')

     
plt.xlabel("Monthly Withdrawl")
plt.axhline(y=0, linestyle='-',linewidth=1, color='black')

plt.show()
```
